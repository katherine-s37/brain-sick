```python
#3倍标准差异常值处理
data1 = data1[data1['delt_v'] < (np.mean(data1['delt_v']) + 3 * np.std(data1['delt_v']))]
data1 = data1[data1['delt_v'] > (np.mean(data1['delt_v']) - 3 * np.std(data1['delt_v']))]
```

```python
data1
```

```python
#二次项拟合
x2 = data1['delt_t'].values
y2 = data1['delt_v'].values
z2 = np.polyfit(x2, y2, 2)
p2 = np.poly1d(z2)
y2_pre = p1(x2)
plt.plot(x2, y2, '.')
plt.plot(np.sort(x2), p1(np.sort(x2)))
print(z2)
```

```python
data1
```

```python
#线形拟合
x1 = data1['delt_t'].values
y1 = data1['delt_v'].values
z1 = np.polyfit(x1, y1, 1)
p1 = np.poly1d(z1)
y1_pre = p1(x1)
plt.plot(x1, y1, '.')
plt.plot(np.sort(x1), p1(np.sort(x1)))
print(z1)
```

```python
#残差计算
res_y2 = y2 - y2_pre
```

```python
res_y2.shape
```

```python
data1['y2_pre'] = y2_pre
data1['res'] = res_y2
```

```python
#结果数据
data1['res'] = abs(data1['res'])
```

```python
data_ = data1.groupby('ID')['res'].sum()
```

```python
data_ = pd.DataFrame(data_).reset_index()
```
