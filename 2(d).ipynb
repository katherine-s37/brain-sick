```python
#导入库
import pandas as pd
import numpy as np
```

```python
#读取数据
data = pd.read_excel('data2d.xlsx')
```

```python
data = data.iloc[:100, :].copy()
```

```python
feat1 = ['HM_volume', 'HM_volume.1', 'HM_volume.2', 'HM_volume.3', 'HM_volume.4', 'HM_volume.5', 'HM_volume.6', 'HM_volume.7', 'HM_volume.8']
feat2 = ['ED_volume', 'ED_volume.1', 'ED_volume.2', 'ED_volume.3', 'ED_volume.4', 'ED_volume.5', 'ED_volume.6', 'ED_volume.7', 'ED_volume.8']
```

```python
#计算欧式距离
dist_list = []
for i in data.iterrows():
    i = i[1]
    hm_value = i[feat1].values
    hm_value = hm_value[hm_value >= 0.0]
    ed_value = i[feat2].values
    ed_value = ed_value[ed_value >= 0.0]
    dist = np.sqrt(np.sum((hm_value - ed_value) ** 2)) / (np.sqrt(np.sum(hm_value ** 2)) * np.sqrt(np.sum(ed_value ** 2)))
    dist_list.append(dist)
```

```python
dist_list
```

```python
data['HM_ED_dist'] = dist_list
```

```python
data.to_csv('result_2d.csv', index=None)
```
